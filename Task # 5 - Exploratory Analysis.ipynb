```python
plt.figure(figsize=(25,15))
plt.title("Company Profit Bar Chart")
sns.barplot(x=df['Sub-Category'],y=df['Profit'])
```

From above graph - Sales of bookcase and tablet dont seems any profitable margin  in fact company facing losses with these two product , but If observed  company have a lot of profit in the sales of copier , but again sales of copier sales is very less we need to increase number of sales of copier

```python
df.groupby('Sub-Category')['Profit','Sales'].sum().plot.bar(color = ['r','b'])
```

```python
plt.figure(figsize = (12,6))
sns.lineplot('Discount', 'Profit',data = df, color = 'y', label = ' Discount')
plt.legend()
```

From the above study , when the profit of state are compared with the discount provided in each state, the states which allowed more discount , went in loss and the above graph we conclude that "THE PROFIT DECREASES WITH THE INCREASE IN DISCOUNT"
